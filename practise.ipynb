```python
#for loop
letters=["a","b","c"]
for letter in letters:
    print(letter)
```

```python
letters=["a","b","c"]
for letter in letters:
    print(letter)
    if letter=="b":
            break
```

```python
letters=["a","b","c","d"]
for letter in letters:
    if letter=="d":
        continue
    print(letter)
        
```
