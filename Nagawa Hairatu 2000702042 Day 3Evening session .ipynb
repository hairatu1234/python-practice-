```python
class Employee:
    def __init__(self,name,empId,salary):
        self.name=name
        self.empId=empId
        self.salary=salary
    def empDetails(self):
        print("Employee Name:",self.name)
        print("Employee Id:",self.empId)
        print("Salary:",self.salary)
obj=Employee("NAGAWA","23Dw","100000000")
obj.empDetails()
    

```

```python
#inheritance
class Animal:
    def __init__(self,name,sound):
        self.name=name
        self.sound=sound
        
class Domestic(Animal):
    def __init__(self,name,sound,character,habitat):
        super().__init__(name,sound)
        self.xtic=character
        self.habitat=habitat
    def display(self):
        print("-------DOMESTIC ANIMAL-------")
        print(f"Animal name:{self.name}")
        print(f"Sound made:{self.sound}")
        print(f"Characteristic :{self.xtic}")
        print(f"habitat :{self.habitat}")
y=Domestic("dog","barks","kept at home","kernel")
y.display()

class Wild(Animal):
    def __init__(self,name,sound,characteristic,habitat):
        super().__init__(name,sound)
        self.xtic=characteristic
        self.habitat=habitat
    def details(self):
        print("-------WILD ANIMAL-------")
        print(f"Animal name:{self.name}")
        print(f"Sound made:{self.sound}")
        print(f"Characteristic :{self.xtic}")
        print(f"habitat :{self.habitat}")
x=Wild("lion","roars","kept in the bush","den")
x.details()
```
